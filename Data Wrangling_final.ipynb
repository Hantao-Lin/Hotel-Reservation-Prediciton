```python
#Import necessary library
import pandas as pd 
```

```python
#Load data
hotel=pd.read_csv('C:/Users/hantao/Desktop/data analysis project/Springboard_Project/Capstone1_Project/hotel_bookings.csv')
weather=pd.read_csv('C:/Users/hantao/Desktop/data analysis project/Springboard_Project/Capstone1_Project/weather_data.csv')
```

```python
#Exam the summaries of both data files
print(hotel.info())
```

```python
print(weather.info())
```

```python
#From hotel data, combine 'arrival_date_year', 'arrival_date_month','arrival_date_day_of_month' to a new column
year=[]
for i in hotel.arrival_date_year:
    year.append(i)
month=[]
for i in hotel.arrival_date_month:
    if i=='July':
        i=7
    if i=='January':
        i=1
    if i=='February':
        i=2
    if i=='March':
        i=3
    if i=='April':
        i=4
    if i=='May':
        i=5
    if i=='June':
        i=6
    if i=='August':
        i=8
    if i=='September':
        i=9
    if i=='October':
        i=10
    if i=='November':
        i=11
    if i=='December':
        i=12
    month.append(i)
day=[]
for i in hotel.arrival_date_day_of_month:
    day.append(i)
date=[]
for i in range(len(month)):
    a=str(month[i])+'/'+str(day[i])+'/'+str(year[i])
    date.append(a)
hotel['Date_time']=date
hotel['Date_time']
```

```python
#Merge two data frames on 'Date_time'
merged=hotel.merge(weather,how='outer')
merged
```

```python
merged.info()
```

```python
#Fill in the nan value in children column
merged['children'].fillna(0,inplace=True)
```

```python
#Fill in the nan value in heat index
t=merged['Average_Temperature']
rh=merged['Relative_Humidity']
heat_index=-42.379+(2.04901523*t)+(10.14333127*rh)-(0.22475541*t*rh)-(6.83783*(10**-3)*t**2)-(5.481717*(10**-2)*rh**2)+(1.22874*(10**-3)*(t**2)*rh)+(8.5282*(10**-4)*t*rh**2)-(1.99*(10**-6)*(t**2)*(rh**2))
merged['Heat_Index'].fillna(heat_index,inplace=True)
```

```python
#Fill in the NA value in country column 
merged['country'].mode()
merged['country'].fillna(value='PRT',inplace=True)
```

```python
#Fill in the NA value in agent
merged['agent'].mode()
merged['agent'].fillna(value=9.0,inplace=True)
merged.info()
```

```python
#Drop the company column
merged.drop('company',axis=1,inplace=True)
```
